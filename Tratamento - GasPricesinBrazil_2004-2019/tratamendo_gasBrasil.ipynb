```python
# Importando a biblioteca
import pandas as pd
```

```python
# Carregando a tabela a atribuindo a ela o nome 'df'
df = pd.read_csv('GasPricesinBrazil_2004-2019.csv', sep=';')
```

```python
df.head(5)
```

```python
# Nesta primeira vizualização já podemos ver uma inconsistencia na tabela,
# que no caso é a coluna 'Unnamed: 0'.
```

```python
df.info()
```

```python
# Analisando mais afundo, percebemos mais alguns erros,
# 1º As colunas ['PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO',
#              'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO'] que são números flutuantes
# estão definidas como string
# 2º As colunas ['DATA INICIAL', 'DATA FINAL'] estão definidas como strings, porém
# podemos definir elas como datetime para uma melhor análise
```

```python
# Vamos começar o tratamendo da tabela eliminando a coluna indesejada
del df['Unnamed: 0']
df.head(5)
```

```python
# Agora vamos alterar o tipo de dado das colunas inconsistentes
colunas = ['DATA INICIAL', 'DATA FINAL']
for coluna in colunas:
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce')
```

```python
colunas = ['PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']
for coluna in colunas:
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
```

```python
# Note que usei laços for para essa operação com o objetivo de
# evitar repetição de código e fazer um código mais performático
```

```python
# Podemos ver que a troca de tipo de dado ocorreu com sucesso, porém
# vemos que algumas colunas perderam algumas linhas, o que indica que
# há caracteres inválidos para aquela coluna/tipo de dado.
# Foi passado para a equipe de engenharia de que os campos nulos
# não seriam úteis para as análises, então, iremos remover todas
# as linhas que apresentam algum campo nulo, para uma melhor precisão.b
df.info()
```

```python
# Por via de curiosidade, vamos conferir os erros
nulos = df['PREÇO MÉDIO DISTRIBUIÇÃO'].isnull()
```

```python
# Os campos determinados nulos, são preenchidos com '-' ou  não apresentam valores numéricos
df[nulos]
```

```python
# Eliminando todas as linhas com valores null ou NaN
df.dropna(inplace=True)
```

```python
# Agora vemos que o tratamento foi realizado com sucesso
# Todas as colunas estão sem valores nulos e com os seus
# tipos de dados correto
df.info()
```

```python
df.to_csv('gasBrasilTratado.csv', sep=';', index=False)
```
