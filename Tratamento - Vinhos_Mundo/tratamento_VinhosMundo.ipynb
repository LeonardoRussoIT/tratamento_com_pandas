```python
# Importando biblioteca Pandas e apelidando de pd
import pandas as pd
```

```python
# Carregando a tabela e tendo a primeira vizualização
df = pd.read_csv("vinhos_mundo.csv", on_bad_lines='skip', sep=',')
df
```

```python
df.info()
```

```python
# Num primeiro momento vimos que os dados da tabela
# estão bastante inconsistentes. Como os tipos de dados
# das colunas estão todos corretos, podemos retirar
# todas as duplicatas e valores nulos da tabela
df = df.dropna()
df = df.drop_duplicates()
```

```python
# A tabela reduziu bastante o tamanho
df.info()
```

```python
# Foi passado para nós da engenharia de dados que só seria
# útil para a equipe de BI as colunas:
# country, points, price, taster_name, title, winery 
# então, vamos retirar as outras

colunas_selecionadas = ['country', 'points', 'price', 'taster_name', 'title', 'winery']
df = df[colunas_selecionadas].copy().reset_index(drop=True)
```

```python
df.columns
```

```python
df.head(4)
```

```python
# Agora vamos alterar o nome das colunas
df.columns = ['PAÍS', 'PONTOS', 'PREÇO', 'NOME DO PROVADOR', 'TÍTULO', 'VINÍCULA']
df.head(3)
```

```python
# Agora vamos alterar a ordem das colunas por pedido da equipe de BI
df = df[['TÍTULO', 'VINÍCULA', 'NOME DO PROVADOR', 'PONTOS', 'PREÇO', 'PAÍS']]
df
```

```python
df.to_csv('vinhos_processado.csv', sep=',', index=False)
```
